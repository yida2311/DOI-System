```python
import os
import json
from tile import Cutter

file_dir = '/media/ldy/7E1CA94545711AE6/OSCC/5x_png'
file_mask_dir = '/media/ldy/7E1CA94545711AE6/OSCC/mask_5x/std_mask'
sample_type = 'seg'

meta_file = '/media/ldy/7E1CA94545711AE6/OSCC/meta/train_val.json'

with open(meta_file, 'r') as f:
    cnt = json.load(f)
```

## train set

```python
slide_list = cnt['train']
save_patch_dir = '/media/ldy/e5a10f4e-18fd-4656-80d8-055bc4078655/OSCC-Tile/5x_1600/train_1600'
save_mask_dir = '/media/ldy/e5a10f4e-18fd-4656-80d8-055bc4078655/OSCC-Tile/5x_1600/train_mask_1600'
save_name = 'tile_info_train_1024.json'

cutter = Cutter(slide_list,
                file_dir,
                file_mask_dir,
                save_patch_dir,
                sample_type=sample_type,
                save_mask_dir=save_mask_dir)

patch_size = 1600
overlap = 400
filter_rate = 0.10

cutter.sample_and_store_patches_png(patch_size,
                                   overlap,
                                   filter_rate,
                                   save_name=save_name)

```
